```python
import requests
from bs4 import BeautifulSoup               #Beautiful Soup是python的一個庫，最主要的功能是從網頁抓取數據。
```

```python
url = "http://news.cnyes.com/Content/20151214/20151214080543107512510.shtml?c=headline"
```

```python
r = requests.get(url)    #所谓 requests.get，就是以 GET 方式发送一个 REQUEST，得到一个 Response 类的结果，保存为 r。
print(r)
```

```python
r.encoding = "UTF-8"                        #將r以UTF-8重新編碼
print(r)
```

```python
soup = BeautifulSoup(r.text,"lxml")         #將r用BeautifulSoup抓取出來並命名為soup，讀入格式使用lxml
print(soup)
```

```python
print(soup.p)                              #抓取soup中有<p>的標籤，但這樣看只有第一個<p>
#print(soup.p.next_sibling)                #因為想不出怎麼從script中抓取正文跟summary，所以改用<p>標籤；缺點是無法抓圖片
```

```python
for sibling in soup.p.next_siblings:       #抓取soup中第二個以後有<p>的標籤
    print(repr(sibling))
```

```python
news_text = soup.select("p")              #將soup中所有<p>標籤選出，命名為news_text
print (repr(news_text))                   #因為news_text是以list讀入，所以用repr檢視所有的list
```

```python
news_text2=news_text[0].get_text()       #將news_text中第1個list轉為text，並命名為news_text2
for i in range(1,len(news_text)):        #將news_text中第2個以後的list轉為text，並存入news_text2
    news_text2=news_text2+news_text[i].get_text()
```

```python
print (news_text2)
```

```python
news_summary = soup.select("p")[0].text  #將soup中第1個<p>標籤轉為text，並命名為news_summary
print (news_summary)                     #鉅亨網中的<p>標籤第1個都是summary，我有比對過
```

```python
import jieba
import jieba.analyse
```

```python
jieba.set_dictionary('dict.txt.big')     #以dict.txt.big建立字典
```

```python
keywords = jieba.analyse.extract_tags(news_text2, topK=10, withWeight=False)     #抓取關鍵字
keywords ="/ ".join(keywords)                                                    #將關鍵字以"/"分隔
print(keywords)
```

```python
print (repr(jieba.analyse.extract_tags(news_text2, topK=10, withWeight=True)))  #unicode-escape在python3中已經默認了，無須轉換
                                                                                #withWeight是關鍵字的權重
```

```python
print (repr(jieba.analyse.textrank(news_text2, topK=10, withWeight=True)))      #.textrank是另一種選關鍵字的方法
```

```python
def index_of_keyword(sentences, keyword):            #定義index_of_keyword函數
    for i, s in enumerate(sentences):                #如果關鍵字在sentences內，就回傳他第一次出現時是在第幾個句子，不再就回傳-1
        if keyword in s:
              return i
    return -1
```

```python
sentences = news_text2.replace("/n","").split(u"。") #將news_text2中的"/n"以。取代，再存入sentences中，並計算有多少個
len(sentences)
```

```python
for sentence in sentences:                           #列出所有的sentences
    print (sentence + "。")
```

```python
index_of_keyword(sentences,u"紫光")                   #回傳紫光第一次出現時是在第幾個句子
```

```python
print (sentences[index_of_keyword(sentences,u"紫光")]+u"。")     #將第7句提出並加上。
```

```python
len(sentences[0])                                               #計算第1句有幾個字
```

```python
s_index = []    
for keyword in keywords[:3]:                                    #抓前3個keywords字元，其實就是抓第一個關鍵字的句子位置
    s_index.append(index_of_keyword(sentences,keyword)) 

s_set = set(s_index)                                            #將他轉成set再轉成list
s_index = list(s_set)
```

```python
summary = ""                                                    #將句子選出並加上"。"
for i in s_index:
    summary += sentences[i]+u"。"
```

```python
print (summary)
```

```python
print (news_summary)
```

```python
print(keywords[:3])
print(s_set)
```

```python
news_key = jieba.analyse.extract_tags(news_text,10,False)           #抓取10個關鍵字
```
